# Eigenvector, Eigen value

As we can see in the next example, eigen value for each dimension is always `0j`.

For each matrix 2x2 we get 2 eigen values.

```python
X = torch.rand((100, 2, 2)) * 100
X[0][0] = torch.tensor([0, 0])
X[0][1] = torch.tensor([0, 0])
for i in range(100):
	x = X[i]
	eigval, eigvec = torch.linalg.eig(x)
	print(x)
	print(eigval)
	print(eigval.imag, eigval.imag == 0)
	print("\n")
	
```
