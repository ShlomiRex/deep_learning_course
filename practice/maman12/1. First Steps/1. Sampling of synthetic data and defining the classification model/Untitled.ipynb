# Generate random 2D points and plot

```python
import sklearn.datasets as skds
# IF we want diffirent randomness we can change the random_state (Remove it)
X, Y = skds.make_blobs(n_samples=100,n_features=2, centers=2, random_state=1)
print(X[:5,:],type(X))
print(Y[:5],type(Y))
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1],
 c=Y, cmap="Greys", edgecolor="black");
```
