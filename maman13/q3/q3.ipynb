```python
import torch
import torchvision
from torch import nn
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
import torchvision.transforms as transforms
import numpy as np
```

```python
BATCH_SIZE = 256
```

```python
transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])
cifar_dataset = torchvision.datasets.CIFAR10(root="./cifar10", train=False, download=True, transform=transform)
torch.utils.data.DataLoader(cifar_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2)
```
